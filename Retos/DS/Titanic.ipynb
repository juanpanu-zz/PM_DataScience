```python
# Mostrar por pantalla el porcentaje de personas que sobrevivieron y murieron
df_survived = df['Survived'].value_counts() / df['Survived'].value_counts().sum()*100
df_survived
```

```python
fig, ax = plt.subplots(figsize=(5, 4), subplot_kw=dict(aspect="equal"))

recipe = ["Muertos",
          "Sobrevivientes"]
data = [x for x in df_survived]
titanic = [x.split()[-1] for x in recipe]

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d})".format(pct, absolute)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),textprops=dict(color="w"))

ax.legend(wedges, titanic,
          title="Titanic",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=12, weight="bold")
ax.set_title("Personas que sobrevivieron en el Titanic",pad=10)
plt.show()
```

```python
#Mostrar por pantalla el porcentaje de personas que sobrevivieron en cada clase.
df_clsvt = df.groupby('Pclass')['Survived'].value_counts() / df['Survived'].value_counts().sum()*100
df_clsvt

```

```python
fig, ax = plt.subplots(figsize=(7, 5), subplot_kw=dict(aspect="equal"))

recipe = ["Clase 1 Sobrevivientes",
          "Clase 1 No Sobrevivientes",
          "Clase 2 No Sobrevivientes",
          "Clase 2 Sobrevivientes",
          "Clase 3 No Sobrevivientes",
          "Clase 3 Sobrevivientes"]

wedges, text = ax.pie(df_clsvt,wedgeprops=dict(width=0.5), startangle=-30)
bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.50)
kw = dict(arrowprops=dict(arrowstyle="-"),bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(recipe[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)
ax.set_title("Personas que sobrevivieron en cada clase",pad=30)
plt.show()
```
